```python
def createEmptyDataFrame(columns, types):
    index = pd.Index([], name="id", dtype=int)
    # specify column name and data type 
    coldefs = zip(columns, types)
    columns = []
    for coldef in coldefs:
        columns.append(coldef)
    # create the dataframe from a dict
    return pd.DataFrame({k: pd.Series(dtype=t) for k, t in columns})
```

```python
colNames = ["seriesid","year","period","value","footnotes"]
colTypes = [str, int, str, float, str]
df = createEmptyDataFrame(colNames, colTypes)
```

```python
code2citytown = {}
citytown2code = {}

file = open("bls-cities-and-towns.raw", "r")

count = 0
for line in file.readlines():
    count += 1
    ignore, code, citytownstate = line.rstrip().split(">", 2)[1].split(" ",2)
    citytown, state = citytownstate.split(",")
    code2citytown[code]=citytown
    citytown2code[citytown]=code
    print("line: {} code: {} citytownstate: {}, citytown: {}, state: {}".format(count, code, citytownstate, citytown, state))

file.close()
```

```python
code2citytown
```

```python
import re
def findCityCode(d, event):
    print("searching for:", event)
    res = []
    for key in d.keys():
        if re.match(event, key):
            res.append(d[key])
    return res
```

```python
city = 'Baltimore'
followedBy = '.*'
search = city + followedBy
print("city: {}, followedBy: {}, search: {}".format(city, followedBy, search))

result = findCityCode(citytown2code, search)
```

```python
for r in result:
    print(r,code2citytown[r])
```
