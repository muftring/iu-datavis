```python
import pandas as pd
```

```python
# read in Top 100 cities, by population
df = pd.read_csv("us-cities-top100-raw.csv", sep=",")
```

```python
df.head()
```

```python
# drop extraneous columns
extraneous = ['Change','2016 land area (english)','2016 land area (metric)','2016 population density (english)','2016 population density (metric)']
df.drop(extraneous, axis=1, inplace=True)
```

```python
df.head()
```

```python
# rename columns
df.rename(columns={'2016 rank': 'rank', 'City': 'city', 'State[5]': 'statename', '2016 estimate': 'population_2016', '2010 Census':'population_2010','Location':'location'}, 
          inplace=True)
```

```python
df.head()
```

```python
# fix location
locations = df.location.values
latitude = []
longitude = []
for location in locations:
    latlon = location.split("/")[1].lstrip().rstrip().replace("°N","").replace("°W","").replace(" ",",-")
    #print(latlon)
    latitude.append(latlon.split(",")[0])
    longitude.append(latlon.split(",")[1])
```

```python
df['latitide'] = latitude
df['longitude'] = longitude
```

```python
df.head()
```

```python
# drop location
df.drop(['location'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# cleanup the city column
cities = []
for city in df.city:
    loc = city.find("[")
    if loc > 0:
        city = city[0:loc]
    #print(city)
    cities.append(city)
```

```python
df.city = cities
```

```python
df.head()
```

```python
# map full state name to abbreviation
state2abbreviation = pd.read_csv("us-states.csv", ",")
```

```python
state2abbreviation.head()
```

```python
state2abbreviation[state2abbreviation.state=='New York'].abbreviation.values[0]
```

```python
abbreviations = []
for state in df.statename:
    abbreviation = state2abbreviation[state2abbreviation.state==state].abbreviation.values[0]
    abbreviations.append(abbreviation)
```

```python
df['state'] = abbreviations
```

```python
df.head()
```

```python
# save as CSV
df.to_csv("us-cities-top100-clean.csv")
```
