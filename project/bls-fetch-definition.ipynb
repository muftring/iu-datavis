The wizard choices are:
1. Seasonally Adjusted or Not Seasonally Adjusted
2. Datatype
3. State
4. Area
5. Supersector
6. Industry

The `SeriesID` is made up of N parts:
- the first two characters are 'SM'
- the next one charater is 'S' (seasonally adjusted) or 'U' (not seasonally adjusted)
- the next two digits indicate the **State** or territory
- the next five digits indicate the metro **Area** in the state
- the next two digits indicate the **Supersector**
- the next six digits indicate the **Industry**
- the final two digits indicate **Datatype**

definitions of each field can be found here https://www.bls.gov/help/def/sm.htm

Examples:
- `SMU01000000000000001`
- `SMS01000000000000001`
