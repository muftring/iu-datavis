```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = pd.read_csv("Quarterly_End_of_Period.csv")
data2 = pd.read_csv("Quarterly.csv")
```

```python
data2
```

```python
df = data1.merge(data2, on='DATE')
```

```python
df
```

```python
df.DATE
```

```python
TNWBSHNO = "Net Worth"
HNOREMV = "Real Estate at Market Value"
TFAABSHNO = "Total Financial Assets"
TLBSHNO = "Total Liabilities"
```

```python
plt.plot(df.TNWBSHNO)
plt.plot(df.TFAABSHNO)
plt.plot(df.TLBSHNO)
plt.plot(df.HNOREMV)
plt.title("U.S. Households and Nonprofit Organizations Balance Sheet\nQuarterly, Not Seasonaly Adjusted")
plt.xlabel('Date')
plt.ylabel('Billions of Dollars')
plt.legend([TNWBSHNO,TFAABSHNO,TLBSHNO,HNOREMV], loc=0)
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.plot(df.TNWBSHNO, label=TNWBSHNO)
ax.plot(df.TFAABSHNO, label=TFAABSHNO)
ax.plot(df.TLBSHNO, label=TLBSHNO)
ax.plot(df.HNOREMV, label=HNOREMV)

plt.title("U.S. Households and Nonprofit Organizations Balance Sheet\nQuarterly, Not Seasonaly Adjusted")
plt.xlabel('Date')
plt.ylabel('Billions of Dollars')

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 1.0, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
